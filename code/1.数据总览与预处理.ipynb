```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# 读取数据
df = pd.read_csv('..\data\Admission_Predict.csv')
df.describe().T.style.background_gradient(axis=0)
```

```python
df.info()
```

```python
# Create a list of DataFrame columns
columns = list(df.columns)
fig, ax = plt.subplots(7, 3, figsize=(15, 45))
plt.subplots_adjust(hspace = 0.5)
for i in range(7) :
    # AX 1
    sns.scatterplot(x=columns[i+1], y='Chance_of_Admit', data=df, hue='Chance_of_Admit', ax=ax[i, 0])
    # Ax 2
    sns.histplot(df[columns[i+1]], bins=5, kde=True, color='purple', ax=ax[i, 1])
    # Ax 3
    sns.boxplot(x=columns[i+1], data=df, ax=ax[i, 2])
    
    
```
